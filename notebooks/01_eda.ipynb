```python
import pandas as pd
```

```python
path_customers = '../data/raw/olist_customers_dataset.csv'
df_customers = pd.read_csv(path_customers)

path_order_items = '../data/raw/olist_order_items_dataset.csv'
df_order_items = pd.read_csv(path_order_items)

path_orders = '../data/raw/olist_orders_dataset.csv'
df_orders = pd.read_csv(path_orders)
```

```python
df_customers.info()
```

```python
df_order_items.info()
```

```python
order_revenue = (
    df_order_items
    .groupby("order_id", as_index = False)["price"]
    .sum()
    .rename(columns = {"price" : "order_revenue"})
)

rev_sum = order_revenue["order_revenue"].sum()
rev_mean = order_revenue["order_revenue"].mean()

print(f"Total revenue is: {rev_sum}")
print(f"Mean of revenue is: {rev_mean}")
```

```python
date_cols = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
    ]
df_orders[date_cols] = df_orders[date_cols].apply(pd.to_datetime)
df_orders.info()
```
